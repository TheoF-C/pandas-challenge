```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

## Player Count

* Display the total number of players


```python
# String formating function for $
def formC(s): return '$' + str(round(s, 2))

# String formating function for %
def formP(s): return str(round(100*s, 2)) + '%'

# Create and dsiplay data frame with total number of players
d = {'Total Players': [len(set(purchase_data["SN"]))]}
display(pd.DataFrame(data=d))
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
# Data frame with summary information for purchasing analysis 
df = pd.DataFrame(data={
         'Number of Unique Items': [len(set(purchase_data["Item Name"]))], 
         'Average Price': [formC(purchase_data["Price"].mean())], 
         'Number of Purchases': [len(purchase_data)], 
         'Total Revenue': [formC(purchase_data["Price"].sum())]}
        )

# Display data frame df
display(df)
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
# Create a copy of purchase_data to trim duplicates
purchase_no_dupes = purchase_data.copy()
purchase_no_dupes.drop_duplicates(subset = ["SN"], keep="first", inplace=True)

gender_count = purchase_no_dupes.groupby('Gender')['Purchase ID'].count().to_frame()
gender_count = gender_count.rename(columns={'Purchase ID': 'Total Count'})

gender_count['Percentage of Players'] = gender_count['Total Count'] / len(purchase_no_dupes)
gender_count['Percentage of Players'] = gender_count['Percentage of Players'].apply(formP)
gender_count.sort_values(by='Total Count', ascending=False, inplace=True)

display(gender_count)
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# Interval list
intervals = [min(purchase_no_dupes['Age']), 9, 14, 19, 24, 29, 34, 39, max(purchase_data['Age'])]

# Find total age count between each interval in intervals
ages = pd.cut(purchase_no_dupes['Age'], intervals, include_lowest=True)
df = pd.DataFrame(data=ages)
ages = [len(df[df['Age'] == age]) for age in sorted(set(ages))]

# Summary data frame with age demographics
df = pd.DataFrame(
    data=[[age, formP(age/len(purchase_no_dupes))] for age in ages],
    index=['<10', '10-14', '15-19', '20-24', '25-29', '30-34', '35-39', '40+'],
    columns=['Total Count', 'Percentage of Players']
)

# Display data frame df
display(df)
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# Find total age count between each interval in intervals
ages = pd.cut(purchase_data['Age'], intervals, include_lowest=True)

# Creates dict with empty list for each age bin
d = {age: [0, 0, 0, 0] for age in sorted(set(ages))}

prices = purchase_data['Price']
for i, k in ages.items():
    d[k][0] += 1 # Sets purchase count
    d[k][2] += prices[i] # Sets total purchase value

for i in purchase_no_dupes['Purchase ID']:
    d[ages[i]][3] += 1; # Finds total individuals in age group

for i in d:
    d[i][1] = formC(d[i][2] / d[i][0]) # Sets average purchase price
    d[i][3] = formC(d[i][2] / d[i][3]) # Sets average total purchase per person
    d[i][2] = formC(d[i][2]) 


# Summary data frame with age purchasing analysis
df = pd.DataFrame(
    data=d.values(),
    index=['<10', '10-14', '15-19', '20-24', '25-29', '30-34', '35-39', '40+'],
    columns=['Purchase Count', 'Average Purchase Price', 'Total Purchase Value', 'Avg Total Purchase per Person']
)
df.index.name = "Age Ranges"

# Display data frame df
display(df)
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Group item name and price to obtain total and average item price
group_by_price = purchase_data.groupby('SN')['Price'].sum()
group_by_sn = purchase_data.groupby('SN')['SN'].count()

# Join group_by_sn and group_by_price
df = pd.concat([group_by_sn, group_by_price], axis=1, join='inner')
df.rename(columns={'Price': 'Total Purchase Value', 'SN': 'Purchase Count'}, inplace=True)

# Calculate average purchase price column
df['Average Purchase Price'] = df['Total Purchase Value'] / df['Purchase Count']

# Sort by total purchase value
df.sort_values(by='Total Purchase Value', ascending=False, inplace=True)

# Apply string formatting
df['Total Purchase Value'] = df['Total Purchase Value'].apply(formC)
df['Average Purchase Price'] = df['Average Purchase Price'].apply(formC)

# Display data frame df
display(df.head(5))
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Group item name and price to obtain item price and total count
item_count = purchase_data.groupby(['Item ID', 'Item Name'])['Item Name'].count()
item_price = purchase_data.groupby(['Item ID', 'Item Name'])['Price'].first()

# Join item_count and item_price
df = pd.concat([item_count, item_price], axis=1, join='inner')
df.rename(columns={'Price': 'Item Price', 'Item Name': 'Purchase Count'}, inplace=True)

# Calculate total purchase value column
df['Total Purchase Value'] = df['Item Price'] * df['Purchase Count']

# Copy data frame for next section
df1 = df.copy()

# Sort by purchase count
df.sort_values(by='Purchase Count', ascending=False, inplace=True)

# Apply string formatting
df['Item Price'] = df['Item Price'].apply(formC)
df['Total Purchase Value'] = df['Total Purchase Value'].apply(formC)

# Display data frame df
display(df.head(5))
```
